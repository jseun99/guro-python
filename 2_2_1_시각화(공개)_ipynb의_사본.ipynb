#3차시 - 인구데이터 시각화
 - 행정안전부(https://jumin.mois.go.kr/)
 - 주민등록 인구통계 - 연령별 인구현황
 - 계, 남녀구분, 연령구분(1세), 만연령구분(0~100세 이상)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
myfile = files.upload()
```

```python
df_pop = pd.read_csv("korea_pop.csv", encoding='cp949', thousands=',')
df_pop.head()
```

```python
#전체 인구 피라미드 그래프
all_man = df_pop.iloc[0,106:207].to_numpy()
all_woman = df_pop.iloc[0, -101:].to_numpy()
x = np.arange(0,101)
print(len(all_man), len(all_woman), len(x))

plt.figure(figsize=(15,10))
plt.barh(x, all_man, label="남자")
plt.barh(x, -all_woman, label='여자')
plt.yticks(np.arange(0,101,5))
plt.legend()
plt.show()
```

```python
#연령대별 남녀차이

plt.figure(figsize=(15,5))
y = all_man - all_woman
plt.bar(x, y, color='r', label='남녀 차이')
plt.legend()
plt.xticks(np.arange(0,101,5))
plt.show()
```

```python
#3 제주도 - 전체 인구 피라미드 그래프
df_jeju = df_pop[df_pop['행정구역'].str.contains('제주')]

jeju_man = df_jeju.iloc[0,106:207].to_numpy()
jeju_woman = df_jeju.iloc[0, -101:].to_numpy()
x = np.arange(0,101)

plt.figure(figsize=(15,10))
plt.barh(x, jeju_man, label="제주 남자")
plt.barh(x, -jeju_woman, label='제주 여자')
plt.yticks(np.arange(0,101,5))
plt.legend()
plt.show()
print('제주남여 차이는 :', np.sum(jeju_man-jeju_woman))
```

```python
#과제4 - 제주 연령대별 남녀차이

plt.figure(figsize=(15,5))
y = jeju_man - jeju_woman
plt.bar(x, y, color='r', label='남녀 차이')
plt.legend()
plt.xticks(np.arange(0,101,5))
plt.show()

max_man = np.argmax(jeju_man-jeju_woman)
max_woman =  np.argmax(jeju_woman-jeju_man)
print('제주 남자가 여자보다 가장 많은 나이는 : ', max_man,'살이고, 인구차이는 :', jeju_man[max_man]-jeju_woman[max_man], '명 입니다.')
print('제주 여자가 남자보다 가장 많은 나이는 : ', max_woman,'살이고, 인구차이는 :', jeju_woman[max_woman]-jeju_man[max_woman], '명 입니다.')
```
